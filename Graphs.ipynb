```python
# This uses the exponential growth formula to predict the running times of these algorithims
# initialize variables
vqe = 0
bf = 0
qaoa = 0
n = 17 # we don't need to start lower since we know the output for this size graph
# Run until brute force takes longer than one of the others
while (bf <= vqe or bf <= qaoa):
    vqe = 1.5229179859161377 * (1.5717)**n
    print("vqe: " + str(vqe))
    qaoa = 4.043205976486206 * (1.5862)**n
    print("qaoa: " + str(qaoa))
    bf = 3.232936018321197e-05 * (1 + 1.2292)**n
    print("br: " + str(bf))
    print("n = " + str(n))
    n+=1

# Reset variables
vqe = 0
bf = 0
qaoa = 0
n = 17 # we don't need to start lower since we know the output for this size graph
# Run until brute force takes longer than vqe to see exactly where it happens
while (bf <= vqe):
    vqe = 1.5229179859161377 * (1.5717)**n
    print("vqe(hours): " + str((vqe/60)/60))
    bf = 3.232936018321197e-05 * (1 + 1.2292)**n
    print("br(hours): " + str((bf/60)/60))
    print("n = " + str(n))
    n+=1
    
# initialize variables
vqe = 0
bf = 0
qaoa = 0
n = 17 # we don't need to start lower since we know the output for this size graph
print("QAOA")
# Run until brute force takes longer than qaoa to see exactly where it happens
while (bf <= qaoa):
    qaoa = 4.043205976486206 * (1.5862)**n
    print("qaoa(hours): " + str((qaoa/60)/60))
    bf = 3.232936018321197e-05 * (1 + 1.2292)**n
    print("br(hours): " + str((bf/60)/60))
    print("n = " + str(n))
    n+=1
```
